## Montando o Grafo

```python
import networkx as nx
import matplotlib.pyplot as plt

#inicia o grafo
G3 = nx.Graph()

#cria um dicionário com as posições [opcional, apenas para visualizão]
pos = {}
pos[1] = (1000, 1500);pos[2] = (1000, 500);pos[3] = (500, 0);pos[4] = (0, 500);pos[5] = (0, 1500);pos[6] = (500, 2000);pos[7] = (1000, 2500);pos[8] = (0, 2500);pos[9] = (500, 3500)

#Adiciona as arestas pela referencia aos nós. O NEtworkX criará os nós automaticamente
G3.add_edge(1, 2)
G3.add_edge(1, 4)
#G3.add_edge(1, 6)


G3.add_edge(2, 5)
G3.add_edge(2, 3)
G3.add_edge(2, 4)

G3.add_edge(3, 4)

G3.add_edge(4, 5)

#G3.add_edge(5, 6)

#G3.add_edge(6, 7)
#G3.add_edge(6, 8)

G3.add_edge(7, 8)
G3.add_edge(7, 9)

G3.add_edge(8, 9)

#desenha o Grafo
options = {
    "font_size": 18,
    "node_size": 500,
    "node_color": "#336699CC",
    "edgecolors": "#002060",
    "edge_color": "#D34817",
    "linewidths": 3,
    "width": 3,
    "with_labels": True
}
plt.figure(figsize = (9,9))
nx.draw_networkx(G3, pos, **options)
```

```python
nx.diameter(G3)
```

```python
#cálculo dos nós
print("Graus dos nós", nx.degree(G3))

#Calculo de distancia
print("\n\n-------------------\n\n")
for i in range(1, 10):
  print(f"A menor distancia entre os nós {i} será")
  for j in range(1, 10):
    print(f"    nó {j}:{nx.shortest_path_length(G3, i, j)}")

#Calculo de centralidade
print("\n\n-------------------\n\n")
print("Centralidade por grau:", nx.degree_centrality(G3))
print("Centralidade por proximidade:", nx.closeness_centrality(G3))
print("Centralidade por intermediação:", nx.betweenness_centrality(G3))
```

```python
import networkx as nx
import matplotlib.pyplot as plt

#inicia o grafo
G3 = nx.Graph()

#Adiciona as arestas pela referencia aos nós. O NEtworkX criará os nós automaticamente
G3.add_edge(1, 2)
G3.add_edge(1, 5)
#G3.add_edge(1, 4)

G3.add_edge(2, 5)
G3.add_edge(2, 4)

G3.add_edge(4, 5)

#desenha o Grafo
options = {
    "font_size": 18,
    "node_size": 500,
    "node_color": "#336699CC",
    "edgecolors": "#002060",
    "edge_color": "#D34817",
    "linewidths": 3,
    "width": 3,
    "with_labels": True
}
plt.figure(figsize = (14,9))
nx.draw_networkx(G3, **options)

print(f"O diametro dessa rede  é {nx.diameter(G3)}")
print(f"O coeficiente médio de clustering dessa rede é {nx.average_clustering(G3)}")
print(f"A distancia média entre os nodos é de  {nx.average_shortest_path_length(G3):.2f}")
print(nx.clustering(G3))
#print(nx.triads.all_triplets(G3))
```
